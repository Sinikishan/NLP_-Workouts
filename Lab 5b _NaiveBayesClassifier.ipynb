**TextBlob** is a Python library for processing textual data.
It provides a simple API for diving into common natural language processing (NLP) tasks such as part-of-speech tagging, noun phrase extraction, sentiment analysis,**classification**, translation, and more.

```python
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
```

Here , we use a simple dataset
where the data is a labeled one

Both data and its sentiment is given as "Pos" or "Neg"

```python
data = [
('I love my country.', 'pos'),
('This is an amazing place!', 'pos'),
('I do not like the smell of this place.', 'neg'),
('I do not like this restaurant', 'neg'),
('I am tired of hearing your nonsense.', 'neg'),
("I always aspire to be like him", 'pos'),
("It's a horrible performance.", "neg")
]
```

```python
import nltk
nltk.download('punkt')
```

We are training the NaiveBayesClassifier over the data ...

Applying the model over the test data

```python
model = NaiveBayesClassifier(data)
model.classify("It's an awesome place!")
```

```python
model.classify("What a dirty place!")
```

This is a simple example to understand , how to use NLTK Package using simple Naive Bayes Classifier over the data.
